Step 3: Load Subject Notes

```python
import os

def load_notes():
    subjects = ["physics", "chemistry", "maths" , "Computer_Science", "english", "biology"]
    data = ""
    for subject in subjects:
        with open(f"{subject}.txt", "r", encoding='utf-8') as f:
            data += f.read() + "\n"
    return data.split("\n\n")  # split by paragraph

documents = load_notes()
print(f"Loaded {len(documents)} academic paragraphs.")
```

 Step 4: Embed Notes with Sentence Transformers

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

doc_embeddings = model.encode(documents)
index = faiss.IndexFlatL2(doc_embeddings[0].shape[0])
index.add(np.array(doc_embeddings))
```

Step 5: Create a Search + Answer System

```python
def get_best_match(question, top_k=1):
    question_embedding = model.encode([question])
    D, I = index.search(np.array(question_embedding), top_k)
    return [documents[i] for i in I[0]]

def ask_bot():
    while True:
        question = input("\nAsk your academic question (or type 'exit'): ")
        if question.lower() == 'exit':
            break
        match = get_best_match(question)[0]
        print("\nAnswer:", match)

ask_bot()
```
